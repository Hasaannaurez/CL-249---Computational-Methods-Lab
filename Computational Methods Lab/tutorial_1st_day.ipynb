```python
def factorial(n):
    if(n==0): return 1
    else: return n*factorial(n-1)
```

```python
def exp(x, terms):
    if(type(x)!=int): 
        print("X in not int")
        return -1
    if(type(terms)!=int): 
        print("Terms in not int")
        return -1
    sum = 0
    for i in range(terms):
        sum = sum + pow(x,i)/factorial(i)
    return sum
```

```python
x, terms = map(int, input().split())
exp(x, terms)
```

```python
def vibonacci_series(n, l, it):
    if(n!=0):
        if(it==1 or it==2):
            l.append(1)
            # print(l)
        else:
            no = random.uniform(0,1)
            if(no<=0.5):
                l.append(abs(l[-1]-l[-2]))
            else:
                l.append(l[-1]+l[-2])
            # print(l)
        vibonacci_series(n-1, l,it+1)

```
