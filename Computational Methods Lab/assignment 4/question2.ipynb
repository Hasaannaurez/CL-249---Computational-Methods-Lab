```python
import numpy as np
```

```python
def lagrange_interpolation(x, y, n, x_p):
    y_p = 0
    for i in range(n):
        term = y[i]
        for j in range(n):
            if(i!=j):
                term = term * (x_p - x[j])/(x[i]-x[j])
        y_p += term
    return round(y_p, 3)
```

```python
# taking input
def take_input_x_y(order):
    x = []
    y = []
    print(f"Enter {order} pairs one by one")
    for i in range(5):
        a, b= map(float, input().split())
        print(f"Entered pair: ({a}, {b})")
        x.append(a)
        y.append(b)

    x = np.array(x)
    y = np.array(y)
    return x, y
```

```python
import matplotlib.pyplot as plt
X = np.array([1, 2.1, 5, 3.3, 4.0])
Y = np.array([2.3, 3.4, 6.2, 5.1, 4.0])
sorted_indices = np.argsort(X)
X = X[sorted_indices]
Y = Y[sorted_indices]
# Target point for xp
xp = 2.5
y_target = lagrange_interpolation(X, Y, len(X),xp)
print(f"f({2.5}) = {y_target}")
# Generate smooth curve for interpolation
x_vals = np.linspace(min(X), max(X), 200)
y_vals = [lagrange_interpolation(X, Y, len(X),xi) for xi in x_vals]

# Plot
plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label="Lagrange Interpolation", color="blue")
plt.scatter(X, Y, color="red", label="Data Points", zorder=5)

#plotting the value in grean at xp=2.5
plt.scatter(xp, y_target, color="green", marker="o", s=100, label=f"Target (x={xp})") 

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Lagrange Interpolation Polynomial")
plt.legend()
plt.grid(True)
plt.show()
```

```python
def take_input_order_value():
    order = int(input()) # m
    value = float(input())

    print(f"Entered No. of Points you want to give = {order}")
    print(f"Entered value = {value}")

    x, y = take_input_x_y(order)
    sorted_indices = np.argsort(X)
    x = x[sorted_indices]
    y = y[sorted_indices]
    # print(x,y)
    print(f"f({value}) = {lagrange_interpolation(x,y, order, value)}")

take_input_order_value()
```
